```python
from telebot.types import InlineKeyboardMarkup, InlineKeyboardButton


user_data = {}

def generate_keyboard():
    keyboard = InlineKeyboardMarkup(row_width=4)
    buttons = [
        "7", "8", "9", "/",
        "4", "5", "6", "*",
        "1", "2", "3", "-",
        "0", ".", "=", "+",
        "C"
    ]
    keyboard.add(*[InlineKeyboardButton(text=symbol, callback_data=symbol) for symbol in buttons])
    return keyboard

@bot.message_handler(commands=['start'])
def start(message):
    user_data[message.chat.id] = ""
    bot.send_message(message.chat.id, "0", reply_markup=generate_keyboard())

@bot.callback_query_handler(func=lambda call: True)
def handle_query(call):
    chat_id = call.message.chat.id
    data = call.data

    if data == "C":
        user_data[chat_id] = ""
        bot.edit_message_text(chat_id=chat_id, message_id=call.message.message_id, text="0", reply_markup=generate_keyboard())
    elif data == "=":
        try:
            result = str(eval(user_data[chat_id]))
            user_data[chat_id] = result
        except Exception:
            result = "–û—à–∏–±–∫–∞"
            user_data[chat_id] = ""
        bot.edit_message_text(chat_id=chat_id, message_id=call.message.message_id, text=result, reply_markup=generate_keyboard())
    else:
        user_data[chat_id] += data
        bot.edit_message_text(chat_id=chat_id, message_id=call.message.message_id, text=user_data[chat_id], reply_markup=generate_keyboard())

if __name__ == "__main__":
    bot.polling(none_stop=True)
```
