## Операторы break, continue
    

Break — ключевое слово break прерывает цикл и передает управление в конец цикла


```python

i = 0
while i < 5:
    a, b = input().split()
    a = int(a)
    b = int(b)
    if (a == 0) and (b == 0):
        break # досрочно завершаем цикл
    print(a * b)
    i += 1
```

Continue — ключевое слово continue прерывает текущую итерацию и передает управление в начало цикла, после чего условие снова проверяется. Если оно истинно, исполняется следующая итерация.

```python
%time
i = 0
while i < 5:
    print('i на входе в цикл',i)
    a, b = input().split()
    a = int(a)
    b = int(b)
    if (a == 0) and (b == 0):
        print('i в случае когда оба равны 0',i)
        break # досрочно завершаем цикл
    if (a == 0) or (b == 0):
        print('i в случае когда хотя бы 1 равен 0',i)
        continue # переходим к следующей итерации
    print(a * b)
    i += 1
    print('i в конце цикла',i)

print('i в конце кода',i)
```

## Цикл for

Цикл for, также называемый циклом с параметром, в языке Питон богат возможностями. В цикле for указывается переменная и множество значений, по которому будет пробегать переменная. Множество значений может быть задано списком, кортежем, строкой или диапазоном.

```python

for i in 2, 3, 5:
    print(i * i)
```

```python
for color in 'red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'violet':
    print('#',  ' color of rainbow is ', color, sep = '')

```

```python
for i in 1, 2, 3, 'one', 'two', 'three':
    print(i)
```

Как правило, циклы for используются либо для повторения какой-либо последовательности действий заданное число раз, либо для изменения значения переменной в цикле от некоторого начального значения до некоторого конечного.

Для повторения цикла некоторое заданное число раз n можно использовать цикл for вместе с функцией range:

```python
for i in range(4):  # равносильно инструкции for i in 0, 1, 2, 3:
    # здесь можно выполнять циклические действия
    print(i)
    print(i ** 2)
# цикл закончился, поскольку закончился блок с отступом
print('Конец цикла')
```

```python

for i in range(10):
    print(i * i)
```

```python
n = int(input())
for i in range(n):
    print('*' * n)
```

Функция range может также принимать не один, а два параметра. Вызов range(a, b) означает, что индексная переменная будеть принимать значения от a до b - 1, то есть первый параметр функции range, вызываемой с двумя параметрами, задает начальное значение индексной переменной, а второй параметр — первое значение, которое индексная переменная принимать не будет. Если же a≥b, то цикл не будет выполнен ни разу. Например, для того, чтобы просуммировать значения чисел от 1 до n можно воспользоваться следующей программой:



Сумма нечетных чисел на отрезке от a до b


```python
a, b = input().split()
a = int(a)
b = int(b)
s = 0
for i in range(a, b + 1):
    if i % 2 == 1:
        s += i
print(s)
```

```python
for i in range(1, 100, 9):
    print(i)
```

Наконец, чтобы организовать цикл, в котором индексная переменная будет уменьшаться, необходимо использовать функцию range с тремя параметрами. Первый параметр задает начальное значение индексной переменной, второй параметр — значение, до которого будет изменяться индексная переменная (не включая его!), а третий параметр — величину изменения индексной переменной. Например, сделать цикл по всем нечетным числам от 1 до 99 можно при помощи функции range(1, 100, 2), а сделать цикл по всем числам от 100 до 1 можно при помощи range(100, 0, -1).

### Добавление if в цикл


Мы можем добавить if в начало цикла, чтобы проверить условия перед продолжением. Вы можете поместить любой условный тест в первую строку и практически любое действие в блоке с отступом после теста. Если условный тест оценивается как True, Python выполняет код, следующий за оператором if. Если результат теста равен False, Python игнорирует код, следующий за оператором if.


```python
magicians  = [ 'Гарри Гудини' , 'Дэвид Копперфилд' , 'Пенн Джилетт' ]
user  =  'Кашперовский'
if  user not in  magicians :
   print(f"Прости, но {user} не волшебник")
```

Отступы в операторах if играют ту же роль, что и в циклах for. Все строки с отступом после оператора if будут выполнены, если тест пройден, и весь блок строк с отступом будет проигнорирован, если тест не пройдет. Вы можете иметь столько строк кода, сколько хотите в блоке, следующем за оператором if.


##  Oператор if-else


```python
magicians  = [ 'Гарри Гудини' , 'Дэвид Копперфилд' , 'Пенн Джилетт' ]
user  =  'Кашперовский'

if user not in magicians:
    print(f"Прости, но {user} не волшебник")
else:
    print(f"{user} настоящий маг!")
```

Если условный тест проходит успешно, выполняется первый блок операторов печати с отступом. Если тест оценивается как False, выполняется блок else.


Решим простую задачку со списками, заодно немного посмотрев на то, как работают всякие циклы и ветвления в этом нашем питоне.

Допустим, у нас есть список чисел и мы хотим его "пофильтровать", оставив только чётные элементы этого списка, сохранив его в новом списке. Есть несколько способов это сделать.

```python
initial = [2, 4, 7335, 753, 2, -17, -22]
```

Способ 1


Это как бы мы сделали в каком-нибудь простом языке типа C с минимумом примочек.


```python
odds1 = []  # пустой список
i = 0
while i < len(initial):
    current = initial[i]
    current_is_odd = ((current % 2) == 1)
    if current_is_odd:
        odds1.append(current)

    i = i + 1
```

```python
odds1
```

Способ 1


Немножко более питонячий, мы, избавляемя от явного i = 0 и i = i + 1, за счёт этого чуть меньше вероятность ошибиться

```python
odds2 = []

for current in initial:
    current_is_odd = ((current % 2) == 1)
    if current_is_odd:
        odds2.append(current)
```

Способ 3
Наиболее питонячий, прикольный, однострочный и молодёжный


```python
odds3 = [current for current in initial if (current % 2 == 1)]
```

```python
print("Способ 1:", odds1)
print("Способ 2:", odds2)
print("Способ 3:", odds3)
```
